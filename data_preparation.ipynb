```python
#methods to get also images
def get_imgs_from_features(row):
    
    specific_dir = main_dir + "/fold" + str(row["fold"]) + "/" + row["slice_file_name"]
    sound_file, sampling_rate = librosa.load(specific_dir, sr = sr, mono=True)
    S = np.abs(librosa.stft(sound_file))
    ft = []
    
    if False:
        mfccs = librosa.feature.mfcc(y=sound_file, sr=sr, n_mfcc=40)
        ft.append(('mfcc', mfccs))
    
    if False:
        spectogram_db = librosa.amplitude_to_db(S, ref = np.max)
        ft.append(('spect', spectogram_db))
    
    if False:
        mel_spectogram = librosa.feature.melspectrogram(y=sound_file, sr=sr)
        ft.append(('mel_spect', mel_spectogram))
    
    if True:
        chroma_stft = librosa.feature.chroma_stft(y=sound_file, sr=sr)
        ft.append(('chroma_stft', chroma_stft))
    
    if True:
        chroma_cqt = librosa.feature.chroma_cqt(y=sound_file, sr=sr)
        ft.append(('chroma_cqt', chroma_cqt))
    
    if True:
        chroma_cens = librosa.feature.chroma_cens(y=sound_file, sr=sr)
        ft.append(('chroma_cens', chroma_cens))
    
    if True:
        spectral_contrast = librosa.feature.spectral_contrast(S=S, sr=sr)
        ft.append(('spectral_contrast', spectral_contrast))
    
    if True:
        y = librosa.effects.harmonic(sound_file)
        if np.sum(np.isnan(y)) == 0:
            tonnetz = librosa.feature.tonnetz(y=y, sr=sr)
            ft.append(('tonnetz', tonnetz))
        
    for f in ft:
        fig = plt.Figure(figsize=(5,5))
        canvas = FigureCanvas(fig)
        cmap = plt.get_cmap('inferno')
        ax = fig.add_subplot(111)
        p = librosa.display.specshow(f[1], x_axis='time', cmap=cmap)
        plt.axis('off')
        plt.savefig(f'/Users/lizzy/Desktop/Universita/cesab/project/img_data/{f[0]}_img/fold{str(row["fold"])}/{row["slice_file_name"][:-4]}.png')
        plt.clf()
    return 
```

```python
def features_matrix(row):
    global count
    global sr
    global mfcc_matrix
    global spectogram_matrix
    global mel_spectogram_matrix
    global chroma_stft_matrix
    global chroma_cqt_matrix
    global chroma_cens_matrix
    global spectral_contrast_matrix
    global tonnetz_matrix
    global rms_matrix
    global spectral_centroid_matrix
    global spectral_bandwidth_matrix
    global spectral_flatness_matrix
    global spectral_rolloff_matrix
    global poly_features_matrix
    global zero_crossing_rate_matrix
    
    specific_dir = f'{main_dir}/fold{str(row["fold"])}/{row["slice_file_name"]}' 
    sound_file, sampling_rate = librosa.load(specific_dir, sr = sr, mono=True)
    # Short-time Fourier transform (STFT)
    S = np.abs(librosa.stft(sound_file))
    
    # Mel-frequency cepstral coefficients (MFCCs)
    mfccs = np.zeros(shape = (40, 173))
    sound_mfcc = librosa.feature.mfcc(y=sound_file, sr=sr, n_mfcc=40)
    if sound_mfcc.shape[1] > 173:
        mfccs[:sound_mfcc.shape[0], :173] = sound_mfcc[:, :173]
    else:
        mfccs[:sound_mfcc.shape[0], :sound_mfcc.shape[1]] = sound_mfcc
    mfcc_matrix[count] = mfccs
    
    # Spectogram
    spectogram = np.zeros(shape = (1025, 173))
    sound_spectogram = librosa.amplitude_to_db(S, ref = np.max)
    if sound_spectogram.shape[1] > 173:
        spectogram[:sound_spectogram.shape[0], :173] = sound_spectogram[:, :173]
    else:
        spectogram[:sound_spectogram.shape[0], :sound_spectogram.shape[1]] = sound_spectogram
    spectogram_matrix[count] = spectogram
    
    # Mel Spectogram: Compute a mel-scaled spectrogram.
    mel_spectogram = np.zeros(shape = (128, 173))
    sound_mel_spectogram = librosa.feature.melspectrogram(y=sound_file, sr=sr)
    if sound_mel_spectogram.shape[1] > 173:
        mel_spectogram[:sound_mel_spectogram.shape[0], :173] = sound_mel_spectogram[:, :173]
    else:
        mel_spectogram[:sound_mel_spectogram.shape[0], :sound_mel_spectogram.shape[1]] = sound_mel_spectogram
    mel_spectogram_matrix[count] = mel_spectogram
    
    # Chroma STFT: Compute a chromagram from a waveform or power spectrogram
    chroma_stft = np.zeros(shape = (12, 173))
    sound_chroma_stft = librosa.feature.chroma_stft(y=sound_file, sr=sr)
    if sound_chroma_stft.shape[1] > 173:
        chroma_stft[:sound_chroma_stft.shape[0], :173] = sound_chroma_stft[:, :173]
    else:
        chroma_stft[:sound_chroma_stft.shape[0], :sound_chroma_stft.shape[1]] = sound_chroma_stft
    chroma_stft_matrix[count] = chroma_stft
    
    # Chroma CQT: Constant-Q chromagram
    chroma_cqt = np.zeros(shape = (12, 173))
    sound_chroma_cqt = librosa.feature.chroma_cqt(y=sound_file, sr=sr)
    if sound_chroma_cqt.shape[1] > 173:
        chroma_cqt[:sound_chroma_cqt.shape[0], :173] = sound_chroma_cqt[:, :173]
    else:
        chroma_cqt[:sound_chroma_cqt.shape[0], :sound_chroma_cqt.shape[1]] = sound_chroma_cqt
    chroma_cqt_matrix[count] = chroma_cqt
    
    # Chroma CENS: Computes the chroma variant “Chroma Energy Normalized” (CENS)
    chroma_cens = np.zeros(shape = (12, 173))
    sound_chroma_cens = librosa.feature.chroma_cens(y=sound_file, sr=sr)
    if sound_chroma_cens.shape[1] > 173:
        chroma_cens[:sound_chroma_cens.shape[0], :173] = sound_chroma_cens[:, :173]
    else:
        chroma_cens[:sound_chroma_cens.shape[0], :sound_chroma_cens.shape[1]] = sound_chroma_cens
    chroma_cens_matrix[count] = chroma_cens

    # Spectral constrast: Compute spectral contrast
    spectral_contrast = np.zeros(shape = (7, 173))
    sound_spectral_contrast = librosa.feature.spectral_contrast(S=S, sr=sr)
    if sound_spectral_contrast.shape[1] > 173:
        spectral_contrast[:sound_spectral_contrast.shape[0], :173] = sound_spectral_contrast[:, :173]
    else:
        spectral_contrast[:sound_spectral_contrast.shape[0], :sound_spectral_contrast.shape[1]] = sound_spectral_contrast
    spectral_contrast_matrix[count] = spectral_contrast
    
    # Tonnetz: Computes the tonal centroid features (tonnetz)
    tonnetz = np.zeros(shape = (6, 173))
    harmonic = librosa.effects.harmonic(sound_file)
    if np.sum(np.isnan(harmonic)) == 0:
        sound_tonnetz = librosa.feature.tonnetz(y=harmonic, sr=sr)
        if sound_tonnetz.shape[1] > 173:
            tonnetz[:sound_tonnetz.shape[0], :173] = sound_tonnetz[:, :173]
        else:
            tonnetz[:sound_tonnetz.shape[0], :sound_tonnetz.shape[1]] = sound_tonnetz
        tonnetz_matrix[count] = tonnetz
    else:
        tonnetz_matrix[count] = np.nan
    
    # RMS: Compute root-mean-square (RMS) value for each frame, either from the audio samples y or from a spectrogram S.
    rms = np.zeros(shape = (173))
    sound_rms = librosa.feature.rms(y=sound_file)[0]
    if sound_rms.shape[0] > 173:
        rms[:173] = sound_rms[:173]
    else:
        rms[:sound_rms.shape[0]] = sound_rms
    rms_matrix[count] = rms
    
    # Spectral centroid: Compute the spectral centroid.
    spectral_centroid = np.zeros(shape = (173))
    sound_spectral_centroid = librosa.feature.spectral_centroid(y=sound_file, sr=sr)[0]
    if sound_spectral_centroid.shape[0] > 173:
        spectral_centroid[:173] = sound_spectral_centroid[:173]
    else:
        spectral_centroid[:sound_spectral_centroid.shape[0]] = sound_spectral_centroid
    spectral_centroid_matrix[count] = spectral_centroid
    
    # Spectral bandwidth: Compute p’th-order spectral bandwidth.
    spectral_bandwidth = np.zeros(shape = (173))
    sound_spectral_bandwidth = librosa.feature.spectral_bandwidth(y=sound_file, sr=sr)[0]
    if sound_spectral_bandwidth.shape[0] > 173:
        spectral_bandwidth[:173] = sound_spectral_bandwidth[:173]
    else:
        spectral_bandwidth[:sound_spectral_bandwidth.shape[0]] = sound_spectral_bandwidth
    spectral_bandwidth_matrix[count] = spectral_bandwidth
    
    # Spectral flatness: Compute spectral flatness
    spectral_flatness = np.zeros(shape = (173))
    sound_spectral_flatness = librosa.feature.spectral_flatness(y=sound_file)[0]
    if sound_spectral_flatness.shape[0] > 173:
        spectral_flatness[:173] = sound_spectral_flatness[:173]
    else:
        spectral_flatness[:sound_spectral_flatness.shape[0]] = sound_spectral_flatness
    spectral_flatness_matrix[count] = spectral_flatness
    
    # Spectral rolloff: Compute roll-off frequency.
    spectral_rolloff = np.zeros(shape = (173))
    sound_spectral_rolloff = librosa.feature.spectral_rolloff(y=sound_file, sr = sr)[0]
    if sound_spectral_rolloff.shape[0] > 173:
        spectral_rolloff[:173] = sound_spectral_rolloff[:173]
    else:
        spectral_rolloff[:sound_spectral_rolloff.shape[0]] = sound_spectral_rolloff
    spectral_rolloff_matrix[count] = spectral_rolloff
    
    # Poly features: Get coefficients of fitting an nth-order polynomial to the columns of a spectrogram.
    poly_features = np.zeros(shape = (2, 173))
    sound_poly_features = librosa.feature.poly_features(y=sound_file, sr = sr)
    if sound_poly_features.shape[1] > 173:
        poly_features[:sound_poly_features.shape[0], :173] = sound_poly_features[:, :173]
    else:
        poly_features[:sound_poly_features.shape[0], :sound_poly_features.shape[1]] = sound_poly_features
    poly_features_matrix[count] = poly_features
    
    # Zero-crossing rate: Compute the zero-crossing rate of an audio time series.
    zero_crossing_rate = np.zeros(shape = (173))
    sound_zero_crossing_rate = librosa.feature.zero_crossing_rate(y=sound_file)[0]
    if sound_zero_crossing_rate.shape[0] > 173:
        zero_crossing_rate[:173] = sound_zero_crossing_rate[:173]
    else:
        zero_crossing_rate[:sound_zero_crossing_rate.shape[0]] = sound_zero_crossing_rate
    zero_crossing_rate_matrix[count] = zero_crossing_rate
    
    count = count + 1
    return 
```
